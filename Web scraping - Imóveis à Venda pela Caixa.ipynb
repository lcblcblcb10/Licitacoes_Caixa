```python
import requests
from lxml import html

# Função para realizar a extração e impressão dos dados de um imóvel específico
def extrair_dados_imovel(numero_imovel):
    # URL da página específica do imóvel
    url = f"https://venda-imoveis.caixa.gov.br/sistema/detalhe-imovel.asp?hdnOrigem=index&hdnimovel={numero_imovel}"

    # Fazendo a requisição HTTP para obter o conteúdo da página
    response = requests.get(url)

    # Verifica se a requisição foi bem sucedida (status code 200)
    if response.status_code == 200:
        # Parseando o conteúdo HTML usando lxml
        tree = html.fromstring(response.content)
        
        # Extraindo o validador
        validador_elements = tree.xpath('//*[@id="dadosImovel"]/div/div/h5/p/text()')
        validador = validador_elements[0].strip() if validador_elements else None
        
        # Se validador for None, continua extraindo os dados
        if validador is None:
            nome_elements = tree.xpath('//*[@id="dadosImovel"]/div/div[1]/h5/text()')
            nome = nome_elements[0].strip() if nome_elements else None

            # Verifica se o nome foi encontrado antes de imprimir os dados
            if nome:
                # Cabeçalhos
                cabecalho_1_elements = tree.xpath('//*[@id="dadosImovel"]/div/div[2]/p/text()[1]')
                cabecalho_1 = cabecalho_1_elements[0].strip() if cabecalho_1_elements else "Não disponível"

                cabecalho_2_elements = tree.xpath('//*[@id="dadosImovel"]/div/div[2]/p/b/text()')
                cabecalho_2 = cabecalho_2_elements[0].strip() if cabecalho_2_elements else "Não disponível"

                cabecalho_3_elements = tree.xpath('//*[@id="dadosImovel"]/div/div[2]/p/text()[2]')
                cabecalho_3 = cabecalho_3_elements[0].strip() if cabecalho_3_elements else "Não disponível"

                # Informações - Parte 1
                informacoes_parte_1 = []
                for i in range(1, 9):  # Itera de 1 a 8
                    pergunta_elemento = tree.xpath(f'//*[@id="dadosImovel"]/div/div[2]/div[1]/p/span[{i}]/text()')
                    pergunta = pergunta_elemento[0].strip() if pergunta_elemento else "Não disponível"

                    resposta_elemento = tree.xpath(f'//*[@id="dadosImovel"]/div/div[2]/div[1]/p/span[{i}]/strong/text()')
                    resposta = resposta_elemento[0].strip() if resposta_elemento else "Não disponível"

                    informacoes_parte_1.append(f"{pergunta} {resposta}")

                # Informações - Parte 2
                informacoes_parte_2 = []
                for i in range(1, 4):  # Itera de 1 a 3
                    pergunta_elemento = tree.xpath(f'//*[@id="dadosImovel"]/div/div[2]/div[2]/p/span[{i}]/text()')
                    pergunta = pergunta_elemento[0].strip() if pergunta_elemento else "Não disponível"

                    resposta_elemento = tree.xpath(f'//*[@id="dadosImovel"]/div/div[2]/div[2]/p/span[{i}]/strong/text()')
                    resposta = resposta_elemento[0].strip() if resposta_elemento else "Não disponível"

                    informacoes_parte_2.append(f"{pergunta} {resposta}")

                # Informações - Parte 3
                informacoes_parte_3 = []
                for i in range(1, 6):  # Itera de 1 a 5
                    resposta_elemento = tree.xpath(f'//*[@id="dadosImovel"]/div/div[3]/span[{i}]/text()')
                    resposta = resposta_elemento[0].strip() if resposta_elemento else "Não disponível"

                    informacoes_parte_3.append(f"{resposta}")

                # Informações - Parte 4
                informacoes_parte_4 = []
                for i in range(1, 3):  # Itera de 1 a 2
                    pergunta_elemento = tree.xpath(f'//*[@id="dadosImovel"]/div/div[3]/p[{i}]/strong/text()')
                    pergunta = pergunta_elemento[0].strip() if pergunta_elemento else "Não disponível"

                    resposta_elemento = tree.xpath(f'//*[@id="dadosImovel"]/div/div[3]/p[{i}]/text()')
                    resposta = resposta_elemento[0].strip() if resposta_elemento else "Não disponível"

                    informacoes_parte_4.append(f"{pergunta} {resposta}")

                # Informações - Parte 5
                informacoes_parte_5 = []
                for i in range(1, 5):  # Itera de 1 a 4
                    resposta_elemento = tree.xpath(f'//*[@id="dadosImovel"]/div/div[3]/p[3]/text()[{i}]')
                    resposta = resposta_elemento[0].strip() if resposta_elemento else "Não disponível"

                    informacoes_parte_5.append(f"Info {i}: {resposta}")

                # Verifica se todos os dados são diferentes de "Não disponível"
                if "Não disponível" not in cabecalho_1:
                    # Imprime os dados do imóvel apenas se todas as informações estiverem disponíveis
                    print(f"Imóvel: {numero_imovel}")
                    print(f"Nome: {nome}")
                    if cabecalho_1 != "Não disponível":
                        print(cabecalho_1)
                    if cabecalho_2 != "Não disponível":
                        print(cabecalho_2)
                    if cabecalho_3 != "Não disponível":
                        print(cabecalho_3)
                    for info in informacoes_parte_1:
                        if "Não disponível" not in info:
                            print(info)
                    for info in informacoes_parte_2:
                        if "Não disponível" not in info:
                            print(info)
                    for info in informacoes_parte_3:
                        if "Não disponível" not in info:
                            print(info)
                    for info in informacoes_parte_4:
                        if "Não disponível" not in info:
                            print(info)
                    for info in informacoes_parte_5:
                        if "Não disponível" not in info:
                            print(info)
                    print("--------------------------------------------")
            else:
                pass  # Não imprime nada se o nome não estiver disponível
        else:
            pass  # Não imprime nada se o validador não estiver disponível
    else:
        pass  # Não imprime nada se a requisição falhar

# Itera sobre os números e extrai os dados dos imóveis
for numero_imovel in range(10002382, 8787715521733):
    extrair_dados_imovel(numero_imovel)
```
